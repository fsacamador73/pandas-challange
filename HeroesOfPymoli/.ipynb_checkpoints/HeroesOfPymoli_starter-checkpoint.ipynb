```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
playerscount = len(purchase_data_df["SN"].unique())
totalplayers_df = pd.DataFrame({"SN":[playerscount]})
totalplayers_df

renamed_df = totalplayers_df.rename(columns = {"SN":"Total Players"})
renamed_df
```

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
UniqueItems = len(purchase_data_df["Item ID"].unique())
AvePrice = round(purchase_data_df["Price"].mean(),2)
CountPurchases = purchase_data_df["Purchase ID"].count()
Revenue = purchase_data_df["Price"].sum()

summary_df = pd.DataFrame({"Number of Unique Items": [UniqueItems],
                           "Average Price": [AvePrice],
                           "Number of Purchases": [CountPurchases],
                           "Total Revenue": [Revenue]
                          })
summary_df

summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
malecount = male_df["Gender"].count()
malepercent = round((malecount / purchase_data_df["SN"].count())*100,2)
#print(malecount)
#print(malepercent)

female_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
femalecount = female_df["Gender"].count()
femalepercent = round((femalecount / purchase_data_df["SN"].count())*100,2)
#print(femalecount)
#print(femalepercent)

others_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]
otherscount = others_df["Gender"].count()
otherspercent = round((otherscount / purchase_data_df["SN"].count())*100,2)
#print(otherscount)
#print(otherspercent)


gender_summary_df = pd.DataFrame({"Total Count": [malecount, femalecount, otherscount],
                                  "Percentage of Players": [malepercent, femalepercent, otherspercent],
                                  "Gender":["Male", "Female", "Other / Non-Disclosed"]
                                 })
gender_summary_df

new_gender_summary_df = gender_summary_df.set_index("Gender")
new_gender_summary_df

new_gender_summary_df["Percentage of Players"] = new_gender_summary_df["Percentage of Players"].map("{:.2f}%".format)
new_gender_summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
male_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
malepurchasecount = male_df["Gender"].count()
malepurchaseaverage = round(male_df["Price"].mean(),2)
malepurchasetotal = male_df["Price"].sum()
uniquemaleusers = len(male_df["SN"].unique())
maleaverageperson = round(malepurchasetotal / uniquemaleusers,2)

#print(malepurchasecount)
#print(malepurchaseaverage)
#print(malepurchasetotal)
#print(uniquemaleusers)
#print(maleaverageperson)

female_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
femalepurchasecount = female_df["Gender"].count()
femalepurchaseaverage = round(female_df["Price"].mean(),2)
femalepurchasetotal = female_df["Price"].sum()
uniquefemaleusers = len(female_df["SN"].unique())
femaleaverageperson = round(femalepurchasetotal / uniquefemaleusers,2)

others_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]
otherspurchasecount = others_df["Gender"].count()
otherspurchaseaverage = round(others_df["Price"].mean(),2)
otherspurchasetotal = others_df["Price"].sum()
uniqueothersusers = len(others_df["SN"].unique())
othersaverageperson = round(otherspurchasetotal / uniqueothersusers,2)


bygender_summary_df = pd.DataFrame({"Purchase Count": [malepurchasecount, femalepurchasecount, otherspurchasecount],
                                    "Average Purchase Price": [malepurchaseaverage,femalepurchaseaverage, otherspurchaseaverage],
                                    "Total Purchase Value": [malepurchasetotal, femalepurchasetotal, otherspurchasetotal],
                                    "Avg Total Purchase per Person":[maleaverageperson, femaleaverageperson, othersaverageperson],
                                    "Gender":["Male", "Female", "Other / Non-Disclosed"]
                                   })
bygender_summary_df

new_bygender_summary_df = bygender_summary_df.set_index("Gender")
new_bygender_summary_df

new_bygender_summary_df["Average Purchase Price"] = new_bygender_summary_df["Average Purchase Price"].map("${:.2f}".format)
new_bygender_summary_df["Total Purchase Value"] = new_bygender_summary_df["Total Purchase Value"].map("${:.2f}".format)
new_bygender_summary_df["Avg Total Purchase per Person"] = new_bygender_summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)
new_bygender_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,44]
bins_labels = ("<10","10-14","15-19","20-24","25-29","30-34","35-39","40+")
#pd.cut(purchase_data_df["Age"], bins, labels = bins_labels)
purchase_data_df["Bins"] = pd.cut(purchase_data_df["Age"], bins, labels = bins_labels)
purchase_data_df

binsbygroup = purchase_data_df.groupby("Bins")
binsbygroup1 = binsbygroup["SN"].count()
#print(binsbygroup1)

percentageplayers = round((binsbygroup["SN"].count() / purchase_data_df["SN"].count())*100,2)
#print(percentageplayers)


bygroup_summary_df = pd.DataFrame({"Total Count": binsbygroup1,
                                   "Percentage of Players": percentageplayers
                                  })
bygroup_summary_df

bygroup_summary_df["Percentage of Players"] = bygroup_summary_df["Percentage of Players"].map("{:.2f}%".format)
bygroup_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9,14,19,24,29,34,39,44]
bins_labels = ("<10","10-14","15-19","20-24","25-29","30-34","35-39","40+")
#pd.cut(purchase_data_df["Age"], bins, labels = bins_labels)
purchase_data_df["Bins"] = pd.cut(purchase_data_df["Age"], bins, labels = bins_labels)
purchase_data_df

binsbygroup = purchase_data_df.groupby("Bins")
purchasecount = binsbygroup["SN"].count()
averagepurchase = round(binsbygroup["Price"].mean(),2)
totalvalue = binsbygroup["Price"].sum()
averageperson = round(totalvalue / purchasecount,2)

byage_summary_df = pd.DataFrame({"Purchase Count": purchasecount,
                                 "Average Purchase Price": averagepurchase,
                                 "Total Purchase Value": totalvalue,
                                 "Avg Total Purchase per Person": averageperson
                                })
byage_summary_df

byage_summary_df["Average Purchase Price"] = byage_summary_df["Average Purchase Price"].map("${:.2f}".format)
byage_summary_df["Total Purchase Value"] = byage_summary_df["Total Purchase Value"].map("${:.2f}".format)
byage_summary_df["Avg Total Purchase per Person"] = byage_summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)
byage_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders = purchase_data_df.groupby(["SN"])["Price"].agg(["count","mean","sum"])
topspenders_df = spenders.sort_values(["sum"], ascending = False)
topspenders_df.head()


renamed_df = topspenders_df.rename(columns = {"count":"Purchase Count","mean": "Average Purchase Price","sum": "Total Purchase Value"})
renamed_df.head()

renamed_df["Average Purchase Price"] = renamed_df["Average Purchase Price"].map("${:.2f}".format)
renamed_df["Total Purchase Value"] = renamed_df["Total Purchase Value"].map("${:.2f}".format)
renamed_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
#popular_df
popularitems = popular_df.groupby(["Item ID", "Item Name"])["Price"].agg(["count","mean","sum"])
mostpopularitems_df = popularitems.sort_values(["count"],ascending = False)
mostpopularitems_df.head()


renamed_df = mostpopularitems_df.rename(columns = {"count":"Purchase Count","mean": "Item Price","sum": "Total Purchase Value"})
renamed_df.head()

renamed_df["Item Price"] = renamed_df["Item Price"].map("${:.2f}".format)
renamed_df["Total Purchase Value"] = renamed_df["Total Purchase Value"].map("${:.2f}".format)
renamed_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popular_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
#popular_df
popularitems = popular_df.groupby(["Item ID", "Item Name"])["Price"].agg(["count","mean","sum"])
mostpopularitems_df = popularitems.sort_values(["sum"],ascending = False)
mostpopularitems_df.head()


renamed_df = mostpopularitems_df.rename(columns = {"count":"Purchase Count","mean": "Item Price","sum": "Total Purchase Value"})
renamed_df.head()

renamed_df["Item Price"] = renamed_df["Item Price"].map("${:.2f}".format)
renamed_df["Total Purchase Value"] = renamed_df["Total Purchase Value"].map("${:.2f}".format)
renamed_df.head()
```
